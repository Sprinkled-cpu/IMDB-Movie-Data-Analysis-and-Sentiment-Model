```python
#display title of movies having runtime >= 180 minutes
#first we have to find the name of the attribute that contains the title of the movie
df.columns
```

```python
df[df['Runtime (Minutes)']>=180]['Title']
```

```python
#finding in which year there was highest average voting
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
#groupby arranges the data here it would show in descending and [] is for specifying on the basis of attributes(condition)mean for average
```

```python
#histogram for the above data output 
sns.histplot(x='Year',y='Votes',data=df)
plt.title("Average Votes by year")
plt.show()
```

```python
#In which Year there was highest average revenue
df.groupby('Revenue (Millions)')['Votes'].mean().sort_values(ascending=False)
```

```python
#Bar Chart for the above data output
sns.barplot(x='Year',y='Revenue (Millions)',data=df)
plt.title("Average Votes by year")
plt.show()
```

```python
#To get average rating for each director
df.groupby('Director')['Rating'].mean().sort_values(ascending=True)
```

```python
#To display Top 10 lengthy movies and their runtime
top10=df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]\
.set_index('Title')
print(top10)
#nlargest function is used here to retrieve largest items w.r.t to amount given
```

```python
#To visualize the above data output
sns.barplot(x='Runtime (Minutes)',y=top10.index,data=top10)
plt.title("Top 10 lengthy moves as per their runtime")
plt.show()
```

```python
#To get no of movies per year
df['Year'].value_counts()
```

```python
#to visualize the above data in a countplot
sns.countplot(x='Year',data=df)
plt.title('No. of Movies in a Year')
plt.show()
```

```python
#To find the most popular movie on the basis of revenue
df[df['Revenue (Millions)'].max()==df['Revenue (Millions)']]['Title']
```
